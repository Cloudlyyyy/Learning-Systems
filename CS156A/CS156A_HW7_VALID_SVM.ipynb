```python
import pandas as pd
import numpy
import math
import random
from sklearn import svm
from tqdm import tqdm
```

```python
df1 = pd.read_table('https://home.work.caltech.edu/data/in.dta', delim_whitespace = True, header = None)
df2 = pd.read_table('https://home.work.caltech.edu/data/out.dta', delim_whitespace = True, header = None)

train = []
val = []
test = []

train_non_lin = []
val_non_lin = []
test_non_lin = []

train_ylist = []
val_ylist = []
test_ylist = []

for row in range(0, 35):
  if row < 25:
    train_coord = (df1.iloc[row][0], df1.iloc[row][1], df1.iloc[row][2])
    train.append(train_coord)
  else:
    val_coord = (df1.iloc[row][0], df1.iloc[row][1], df1.iloc[row][2])
    val.append(val_coord)
```

```python
# preform non-linear transformation (train) for k = 3 to 7

for k in range(3,8):
  train_non_lin_k = []
  for x in range(len(train)):
    x1 = train[x][0]
    x2 = train[x][1]
    y = train[x][2]

    train_ylist.append(y)

    non_lin_alg = [1, x1, x2, x1**2, x2**2, x1*x2, abs(x1-x2), abs(x1+x2)]
    non_lin_alg_k = non_lin_alg[0:k+1]
    train_non_lin_k.append(non_lin_alg_k)

  train_non_lin.append(train_non_lin_k)

print(train_non_lin)
```

```python
# preform non-linear transformation (train) for k = 3 to 7

for k in range(3,8):
  val_non_lin_k = []
  for x in range(len(val)):
    x1 = val[x][0]
    x2 = val[x][1]
    y = val[x][2]

    val_ylist.append(y)

    non_lin_alg = [1, x1, x2, x1**2, x2**2, x1*x2, abs(x1-x2), abs(x1+x2)]
    non_lin_alg_k = non_lin_alg[0:k+1]
    val_non_lin_k.append(non_lin_alg_k)

  val_non_lin.append(val_non_lin_k)

print(val_non_lin)
```

```python
# preform non-linear transformation (test) for k = 3 to 7

for k in range(3,8):
  test_non_lin_k = []
  for x in range(len(df2)):
    x1 = df2.iloc[x][0]
    x2 = df2.iloc[x][1]
    y = df2.iloc[x][2]

    test_ylist.append(y)

    non_lin_alg = [1, x1, x2, x1**2, x2**2, x1*x2, abs(x1-x2), abs(x1+x2)]
    non_lin_alg_k = non_lin_alg[0:k+1]
    test_non_lin_k.append(non_lin_alg_k)

  test_non_lin.append(test_non_lin_k)

print(test_non_lin)
```

```python
# run linear regression for k = 3 to 7

for k in range(3,8):
  pseudoin = numpy.linalg.pinv(train_non_lin[k-3])
  for i in range(0, len(train_ylist), 25):
    sub_train_ylist = train_ylist[i:i+25]
    weight = numpy.matmul(pseudoin, sub_train_ylist)

  # val error for k = 3 to 7
  misclassified = 0
  for i in range(len(val_non_lin[k-3])):
    if (numpy.sign(numpy.matmul(weight, val_non_lin[k-3][i])) != val_ylist[i]):
      misclassified += 1

  print(f"val error for k = {k}: {misclassified / len(val_non_lin[k-3])}")
```

```python
def perceptron(N):
    # create two random points
    pt_1 = [random.uniform(-1,1), random.uniform(-1,1)]
    pt_2 = [random.uniform(-1,1), random.uniform(-1,1)]

    # calculate slope using formula = (y2 - y1) / (x2 - x1)
    m = (pt_2[1] - pt_1[1]) / (pt_2[0] - pt_1[0])

    # calculate y-int = b = y1 - mx1
    b = pt_1[1] - (m * pt_1[0])

    lst_scores = []

    for i in range(N):

        # create random point
        x_n = random.uniform(-1,1)
        y_n = random.uniform(-1,1)

        # initializing
        tup = (1, x_n, y_n)
        score = 0

        # scoring
        if ((m * x_n) + b) > y_n:
            score = 1
        if ((m * x_n) + b) < y_n:
            score = -1

        # add score to list
        lst_scores.append((tup, score))

    same_point = lst_scores[0][1]
    counter = 0

    for i in range(N):
      if lst_scores[i][1] == same_point:
        counter +=1
    if counter == N:
        return perceptron(N)

    weight = [0,0,0]
    iteration = 0

    while True:

        # get missclassified points

        miss_point = []

        for point,score in lst_scores:

            # dot product of point and weight

            value = int(numpy.sign((point[0] * weight[0]) + (point[1] * weight[1]) + (point[2] * weight[2])))

            if value != score:
                miss_point.append((point, score))

        if len(miss_point) > 0:

            # updating weight

            iteration += 1
            p,s = random.choice(miss_point)
            weight = [(weight[0] + (s * p[0])), (weight[1] + (s * p[1])), (weight[2] + (s * p[2]))]
        else:
            return iteration, weight, m, b, lst_scores

def SVM(lst_scores, m, b):
  x = []
  y = []
  num_rand_point = 100
  dis = 0


  for i in range(len(lst_scores)):
      x_temp = lst_scores[i][0]
      y_temp = lst_scores[i][1]
      x.append(x_temp)
      y.append(y_temp)

  clf = svm.SVC(C = 100000, kernel = 'linear')
  clf.fit(x, y)

  for ele in range(num_rand_point):
    x_n = random.uniform(-1,1)
    y_n = random.uniform(-1,1)
    point = (1, x_n, y_n)
    if ((m * x_n) + b) > y_n:
        score = 1
    if ((m * x_n) + b) < y_n:
        score = -1

    if clf.predict([point]) != score:
      dis += 1

  return dis/num_rand_point

def answers():
    num_iter = []
    list_disagreement = []
    num_point = 10
    num_rand_point = 100
    for i in range(100):
        iteration, weight, m, b, lst_scores = perceptron(num_point)
        num_iter.append(iteration)
        disagree = 0
        for ele in range(num_rand_point):
            x_n = random.uniform(-1,1)
            y_n = random.uniform(-1,1)
            point = (1, x_n, y_n)
            score = 0
            if ((m * x_n) + b) > y_n:
                score = 1
            if ((m * x_n) + b) < y_n:
                score = -1
            value = int(numpy.sign((point[0] * weight[0]) + (point[1] * weight[1]) + (point[2] * weight[2])))
            if value != score:
                disagree += 1
        list_disagreement.append(disagree / num_rand_point)
        error = SVM(lst_scores, m, b)
    return sum(num_iter) / len(num_iter) , sum(list_disagreement) / len(list_disagreement) , error

better_prob = 0;

for i in tqdm(range(1000)):
  num_iter, pla_error, svm_error = answers()
  if svm_error < pla_error:
    better_prob += 1
```
